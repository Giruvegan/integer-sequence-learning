```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
plt.rcParams["figure.figsize"] = [8, 6]
```

```python
train_set = pd.read_csv('train.csv')
test_set = pd.read_csv('test.csv')
print 'train_set shape: '
print train_set.shape
print 'test_set shape: ' 
print test_set.shape
```

```python
train_set['Sequence'] = train_set['Sequence'].apply(lambda x: x.split(','))
train_set['length'] = pd.Series(list(train_set['Sequence'].apply(lambda x: len(x))), index=train_set.index)
train_set['flag'] = 'train'
test_set['Sequence'] = test_set['Sequence'].apply(lambda x: x.split(','))
test_set['length'] = pd.Series(list(test_set['Sequence'].apply(lambda x: len(x))), index=test_set.index)
test_set['flag'] = 'test'
df = pd.concat([train_set, test_set])
df = df.sort_values('Id')
```

```python
def df_2_dict(df, key_name, value_name):
    return dict(zip(list(df[key_name]), list(df[value_name])))
```

```python
def plot_sequence(id, save=False):
    value = list(df[df['Id']==id]['Sequence'])[0]
    if save:
        plt.ioff()
    fig = plt.figure()
    plt.plot(np.arange(len(value)), np.asarray(value))
    if not save:
        plt.show()
    else:
        path = os.path.join('/data/zhe.gao/kaggle/integer_sequence_learning', 'plots')
        if not os.path.exists(path):
            os.makedirs(path)
        if not os.path.exists(os.path.join(path, str(id)+'.png')):
            plt.savefig(os.path.join(path, str(id)+'.png'))
    plt.close(fig)
```
